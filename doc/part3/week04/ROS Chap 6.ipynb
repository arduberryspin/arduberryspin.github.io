# ROS 도구

여기서는 크게 5가지 보조 도구에 대해서 설명합니다.

* RViz : 3D Visualization Tool
* rqt_bag : Logging and Visualization Sensor Data, rosbag gui tool
* rqt_plot : Data Plot Tool
* rqt_graph : 노드와 메시지간의 상관 관계를 그래프로 나타내는 툴
* rqt : QT 기반의 ROS GUI 개발 툴

이들은 ROS에 핵심 도구는 아니지만, ROS를 수행하는데 지원을 해주는 것들 입니다.

## RViz : 3차원 시각화 도구

RViz는 말 그대로 ROS에서 수집되는 데이터들을 3차원의 시각화하여 확인 할수 있게 해줍니다. 예를 들어, 

* 레이저 레인지 파인더(LRF: Laser Range Finder)의 센서로부터의 장애물과의 거 리값, 
* Kinect 및 Xtion과 같은 3차원 거리 센서의 포인트 클라우드 데이터 (PCD: Point Cloud Data), 
* 카메라로부터 취득한 컬러 이미지 값 

등을 실시간으로 표시할 수 있습니다.

또한,

* 사용자가 지정한 폴리곤을 이용하여 각종 표시를 지원하고,
* Interactive Markers 를 이용하여 사용자 노드로부터의 명령 및 데이터를 수신하여 상호호환적인 움직임을 지원하고,
* URDF(Unified Robot Description Format)을 이용하여 로봇을 3차원 모델로 나타내고, 여기에서 정의된 각 서브모델들에 대해서 그 자유도에 따라 이동 및 구동이 가능하도록 할 수 있는 기능

을 지원합니다.


### RViz 설치 및 실행

```
sudo apt-get install ros-indigo-rviz
```

실행명령은 다음과 같습니다.

```
rosrun rviz rviz
```

## rqt : QT로 개발된 ROS 관리 도구

rqt_graph, rqt_plot, rqt_bag등은 rqt의 plug-in으로써 rqt가 설치되어 있어야 실행이 가능합니다. 그러므로, rqt를 먼저 설치하도록 하겠습니다.

* rqt 설치

Rviz와 마찬가지로 Desktop-Full Install로 설치하지 않았다면, 아래의 명령어를 이용하여 설치하여야 합니다.

```
sudo apt-get install ros-indigo-rqt ros-indigo-rqt-common-plugins
```

위와 같이 설치할 수 있지만, 추가적으로 설치해야할 것이 있습니다. 이는 rqt_plot 및 rqt_graph의 실행을 원활하게 해주는 도구로써, PyQtGraph라는 Package를 설치하면 됩니다.

책에서는 

http://www.pyqtgraph.org/downloads/python-pyqtgraph-0.9.8-1_all.deb

의 파일을 받아서 설치하도록 하고 있습니다. 

그 후 ```rqt_graph```를 실행하여, 설정(오른쪽 상단에 있는 옵션을 의미하는 기어모양의 아이콘)을 클릭하여, PyQtGraph를 사용하겠다고 설정해주면 됩니다.

* rqt 실행

```
rqt
```

를 Command Prompt에 입력해서 실행할 수 있습니다.

rqt를 최초 실행시 아무것도 나오지 않는데, 이는 사용하려는 플러그인을 선택하지 않아서 그런 것 입니다. 사용하려는 플러그인을 'Plugins'라는 메뉴에서 선택하면 됩니다. 'Running'메뉴는 현재 실행되는 플러그인의 종류 및 그들을 강제로 종료할 수 있는 기능을 제공하는 메뉴이며, 'Perspectives'는 현재 구동중인 플러그인을 다음에도 rqt실행시에 동일하게 띄우고 싶을때 사용할 수 있는 메뉴입니다.

* rqt 플러그인

플러그인 메뉴에 있는 플러그인들 몇개에 대해서 간단하게 설명을하면,

** Action

*** Action Type Browser

Action 타입의 데이터 구조를 확인하는 플러그인이라고 합니다.

** Configuration

*** Dynamic Reconfigure

노드들에서 제공하는 설정값 변경을 위한 GUI 설정값 변경 플러그인 

*** Launch 

roslaunch 의 GUI 플러그인, ROS런치의 이름 및 구성이 생각안날때 매우 유용하다.

** Introspection

Node Graph | 구동중인 모든 노드들의 관계도 및 메시지의 흐름을 확인할 수 있는 그래프 뷰 형식의 플러그인

Package Graph | 노드의 의존 관계를 표시하는 그래프 뷰 형식의 플러그인

Process Monitor | 현재 실행중인 노드들의 CPU사용률, 메모리사용륭, 스레드수 등을 확인 가능하다.

** Logging

Bag | ROS 데이터 로깅 관련 플러그인

Console | 노드들에서 발생되는 경고(Warning), 에러(Error) 등의 메시지들을 한 화면에서 확인 가능한 플러그인

Logger Level | ROS의 Debug, Info, Warn, Error, Fatal 로거 정보를 선택하여 표시할 수 있는 GUI 툴

** Miscellaneous Tools

Python Console | 파이썬 콘솔 화면 플러그인

Shell | 쉘(shell)을 구동하는 플러그인 Web | 웹 브라우저를 구동하는 플러그인

** Robot

사용하는 로봇에 따라 계기판(dashboard) 등의 플러그인을 이곳에 추가하면 된다.

** Robot Tools

Controller Manager | 컨트롤러 제어에 필요한 플로그인

Diagnostic Viewer | 로봇 디바이스 및 에러 확인 플러그인

Moveit! Monitor | 로봇 팔 계획에 사용되는 Moveit! 데이터를 확인하는 플러그인

Robot Steering | 로봇 조정 GUI 툴, 원격 조정에서 이 GUI 툴을 이용하여 로봇 조종하기에 유용함.

Runtime Monitor | 실시간으로 노드들에서 발생되는 에러 및 경고를 확인가능한 플러그인

** Services

Service Caller | 구동중인 서비스 서버에 접속하여 서비스를 요청하는 GUI 플러그인, 서비스 테스트에 유용하다.

Service Type Browser | 서비스 타입의 데이터 구조를 확인하는 플러그인

** Topics

Easy Message Publisher | 토픽을 GUI 환경에서 발행가능 한 플러그인

Topic Publisher | 토픽을 생성하여 발행가능한 GUI 플러그인, 토픽 테스트에 매우 유용하다. 

Topic Type Browser | 토픽 타입의 데이터 구조 확인 플러그인, 토픽 타입 확인에 매우 유용하다.

Topic Monitor | 현재 사용중인 토픽을 나열, 그 중에서 사용자가 선택한 토픽의 정보를 확인하는 플러그인

** Visualization

Image View | 카메라의 영상 데이터를 확인 가능한 플러그인, 간단한 카메라 데이터 테스트에 유용하다.

Navigation Viewer | 로봇 네비게이션의 위치 및 목표지점 확인하는 플러그인 

Plot | 2차원 데이터 플롯 GUI 플러그인, 2차원 데이터의 도식화에 매우 유용하다. 

Pose View | 현재 tf의 위치 및 모델의 위치 표시 플러그인

RViz | 3차원 시각화 툴인 RViz 플러그인

TF Tree | tf 관계를 트리로 나타내는 그래프 뷰 형식의 플러그인

### rqt_plot

rqt_plot은 rqt의 플러그인의 일종입니다. plot이라는 명칭이 들어있는 것처럼 이는 ROS로부터 메시지를 받아서 이를 좌표로 뿌리는 것과 같은 작업을 수행합니다.

여기서는 실습으로 turtlesim 노드 pose 메시지의 x 좌표와 y좌표를 좌표에 받아서 이를 rqt_plot화면에 출력해보겠습니다.

실행순서는 다음과 같습니다. 

```
$ rosrun turtlesim turtlesim_node
$ rqt_plot /turtle1/pose/
$ rosrun turtlesim turtle_teleop_key
```

### rqt_bag

메시지 기록을 시각화하는 rqt의 plugin입니다. 5장에서 설명한 rosbag에서 설명한 자료들을 시각화하여 표출해준다고 합니다.

책에서는 카메라를 가지고 있는 실제의 turtle machine에서 발생하는 메시지들을 특정 bag파일에 저장하고 이를 시각화하여 보여주는 예제를 보여주고 있습니다.

여기서는 그러한 내용에 대해서는 설명으로 대체하겠습니다.

### rqt_graph

각 node간에 어떤 데이터들이 전송되는지 어떤 흐름을 보이는지 체크할 수 있는 plugin이라고 합니다.
사용방법은 매우 간단하여, rqt를 실행시키고 plugins 메뉴에서 ROS Graph를 선택하면 된다고 합니다.

### Image View

카메라의 이미지 데이터를 표시하는 플러그인입니다. 이미지 처리 프로세스는 아니지만, 단순히 영상을 확인하는 용도로는 매우 간단하기에 유용하다고 합니다.

테스트를 위해서는 다음과 같은 패키지를 설치하셔야 합니다.

```
$ sudo apt−get install ros−indigo−uvc−camera
```

책에서는 컴퓨터의 USB에 Camera를 연결하여 이를 통한 영상을 보여주는 샘플을 보여주고 있습니다만, 여기서는 docker를 이용한 것이므로, 설명으로 대체하겠습니다.

책에서는 USB CAM을 컴퓨터의 USB에 연결하고, 아래의 명령어로 uvc_camera 패키지의 uvc_camera_node 노드를 구동하시면,

```
$ rosrun uvc_camera uvc_camera_node
```

후

```
$ rqt
```

를 실행하여 plugins에 Image View 선택 후 Image View화면의 메시지 선택란에 /image_raw라고 입력하면 된다고 합니다.
