###1. Ubuntu 14.04 LTS + ROS

* ###먼저 가급적 ntp 설정 </br></br>

* ###패키지 다운로드를 위해 공개키를 추가
  * <b><font color="blue">$ wget https://raw.githubusercontent.com/ros/rosdistro/master/ros.key -O - | sudo apt-key add -</font></b>

  - 이 명령어로 데스크톱용 기본적인 ROS 패키지들을 설치하게 된다. 여기에는 ROS, rqt, rviz, 로봇 관련 라이브러리, 시뮬레이션, 네이게이션 등이 포함되어 있다.

  - <font color="blue"><b>$ sudo apt-get install ros-indigo-desktop-full</b> </font>

- ###rqt 패키지 추가 설치
  - rqt관련 모든 패키지를 설치하는것이 여러모로 편하고, 다양 한 rqt 플러그인을 사용가능해진다.
  - <b><font color="blue">$ sudo apt-get install ros-indigo-rqt-*</font></b>

- ###다음 명령어로 ros관련 패키지 검색 가능

- ### ROS 개발 환경 구축 (IDE)
  - 통합 개발 환경 (Integrated Development Environment, IDE)
  <br/><br/>
  - QtCreator 설치
    - <b><font color="blue">$ sudo apt-get install qtcreator</font></b>
  <br/><br/>
  
  - QtCreator 구동
    - <b><font color="blue">$ qtcreator</font></b>
    <br/><br/>
    
  - ROS 패키지를 프로젝트로 불러오기
    - QtCreator는 기본적으로 CMakeLists.txt 를 사용하고 있고, ROS 패키지도 CMakeLists.txt 기반이기에<br/> 
    단순히 아래와 화면의 OpenProject 버튼을 클릭하여 해당 ROS 패키지의 CMakeLists.txt 를 선택하면 손쉽게 프로젝트로 불러올 수 있다.
    <br/><br/>
    
    - 컴파일의 경우에는 단순히 Ctrl + b 로 컴파일 하면 catkin_make 가 실행된다. <br/>단, 빌드 관련은 해당 패키지와 같은 위치의 폴더에 새로운 폴더로 생성된다. 예를 들어 tms_rp_action 이라는 패키지를 컴파일하면 build-tms_rp_action-Desktop-Default 이라는 폴더에 빌드 관련이 모두 놓이게 된다. <br/><br/>
    즉, 원래는 ~/catkin_ws/build 와 ~/catkin_ws/devel 에 보관되어야 할 파일들이 따 로 컴파일 되어 새로운 장소에 놓이게 되므로 실행을 위해서는 나중에 다시 한번 catkin_make 를 해줘야 한다. 이는 매번 해줄 필요는 없고 개발 도중에는 QtCreator 에서 개발, 디버깅한 후에 완료되어 실행할 때만 따로 catkin_make 를 해주면 된다.

###2. Raspbian + ROS

- Rasbian 에서는 기본적인 ROS 설치 방법이나 catkin make 가 바이너리로 제공되지 않음....<br/><br/>

- <b><font color="blue">$ sudo sh -c 'echo "deb http://packages.ros.org/ros/ubuntu wheezy main" > /etc/apt/sources.list.d/ros-latest. list'</font></b>

- <b><font color="blue">$ wget https://raw.githubusercontent.com/ros/rosdistro/master/ros.key -O - | sudo apt-key add - </font></b>

- <b><font color="blue">$ sudo apt-get update</font></b>

- <b><font color="blue">$ sudo apt-get update</font></b>

- <b><font color="blue">$ sudo apt-get upgrade</font></b>

- <b><font color="blue">$ sudo apt-get install python-setuptools</font></b>

- <b><font color="blue">$ sudo easy_install pip</font></b>

- <b><font color="blue">$ sudo pip install -U rosdep rosinstall_generator wstool rosinstall</font></b>

- <b><font color="blue">$ sudo rosdep init</font></b>

- <b><font color="blue">$ rosdep update</font></b>

- <b><font color="blue">$ mkdir ~/catkin_ws</font></b>

- <b><font color="blue">$ cd ~/catkin_ws</font></b>

- <b><font color="blue">$ rosinstall_generator ros_comm --rosdistro indigo --deps --wet-only --exclude roslisp --tar > indigo -ros_comm-wet.rosinstall</font></b>

- <b><font color="blue">$ wstool init -j8 src indigo-ros_comm-wet.rosinstall</font></b>

- <b><font color="blue">$ mkdir ~/catkin_ws/external_src</font></b>

- <b><font color="blue">$ sudo apt-get install checkinstall cmake</font></b>

- <b><font color="blue">$ cd ~/catkin_ws/external_src</font></b>

- <b><font color="blue">$ sudo apt-get install libboost-system-dev libboost-thread-dev</font></b>

- <b><font color="blue">$ git clone https://github.com/ros/console_bridge.git</font></b>

- <b><font color="blue">$ cd console_bridge</font></b>

- <b><font color="blue">$ cmake .</font></b>

- <b><font color="blue">$ sudo checkinstall make install</font></b>
  - checkinstall 실행 후 EOF 구문에서 엔터키를 누르고, 2 번쩨 항목에서 console-bridge 부분을 libconsole-bridge-dev 로 바꿔준다.
<br/><br/>

- <b><font color="blue">$ cd ~/catkin_ws/external_src</font></b>

- <b><font color="blue">$ wget http://archive.raspbian.org/raspbian/pool/main/l/lz4/liblz4-1_0.0~r122-2_armhf.deb</font></b>

- <b><font color="blue">$ wget http://archive.raspbian.org/raspbian/pool/main/l/lz4/liblz4-dev_0.0~r122-2_armhf.deb</font></b>

- <b><font color="blue">$ sudo dpkg -i liblz4-1_0.0~r122-2_armhf.deb liblz4-dev_0.0~r122-2_armhf.deb</font></b>

- <b><font color="blue">$ cd ~/catkin_ws</font></b>

- <b><font color="blue">$ rosdep install --from-paths src --ignore-src --rosdistro indigo -y -r --os=debian:wheezy</font></b>

- <b><font color="blue">$ sudo ./src/catkin/bin/catkin_make_isolated --install -DCMAKE_BUILD_TYPE=Release --install-space /opt/ros/indigo</font></b>

- <b><font color="blue">$ source /opt/ros/indigo/setup.bash</font></b>
<br/><br/>

- 마지막으로 catkin_make와 roscore가 정상동작 하는지 확인.
  - <b><font color="blue">$ cd ~/catkin_ws/</font></b>
  
  - <b><font color="blue">$ catkin_make</font></b>
  
  - <b><font color="blue">$ roscore</font></b>
