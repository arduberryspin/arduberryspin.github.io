## 아두이노의 등장과 배경  

-- 와이어링  

* IDII의 학생인 에르난도 바리간이 2003년부터 개발하여 2004년에 최초 버전이 릴리즈된 도구  
* IDII는 이탈리아 북부의 이브레아에서 2001년 개교한 국제적인 인터랙션 디자인 연구교육기관  
* 새계적으로 많이 쓰이던 AVR마이크로컴퓨터 보드 및 거기에 프로그램을 올릴 수 있도록 하는 IDE와 라이브러리를 제공함. 언어로는 프로세싱과 유사한 언어 채택  
* 그 후 남미의 콜롬비아 로스 안데스 대학으로 옮겨 개발을 계속함  

-- 아두이노  

* 2004년부터 개발 2005년에 최초 보드가 제조  
* 와이어링은 사용하기 쉬운 도구였지만, 가격이 약간 비싸서(10만원) 교육용으로 사용하기에는 무리  
* 마시모 밴지([손에 잡히는 아두이노]의 저자) 팀이 와이어링을 기초로 최대한 간략하게 만든 버젼  
* 아두이노의 장점은 여타 다른 장비들과 다르게 하드웨어 그 자체에 대한 오픈 뿐만이 아니라 개발환경 자체도 제공한다는 점이 매력  
* 즉, 직접 만들기 위한 여러가지 필요한 도구들을 바로 제공한다는 점

프로토타이핑?

* 책의 저자는 우리가 원하는 형태의 물건을 만들기 위해서 자주 그리고 계속 만들어볼 것을 제안하고 있다.

* 계속 만들어서 발전시켜나가므로써, 문제점 및 제작하는 아이템에 대한 개선 방향등에 대한 아이디어들이 생길 수 있다는 것이다.

* 듀얼사이클론 방식의 청소기를 개발한 제임스 다이슨은 이러한 과정을 5127회 반복했다고 한다.
